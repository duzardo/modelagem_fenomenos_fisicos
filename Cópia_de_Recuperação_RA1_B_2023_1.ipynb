#Questão 1
A seguir estão listados três limites. Sua tarefa será encontar o valor destes limites de duas formas: algébricamente e gráficamente.

a) Para a solução algébrica, você deve explicitar todas as passagens necessárias em latex, em uma célula de texto.

b) Para solucionar graficamente deverá usar uma tabela de aproximação pela esquerda e pela direita. Lembre-se de plotar o gráfico, os pontos usados para a aproximação e de imprimir a tabela no terminal. 


1. $$\lim_{{x \to 0}} \frac{1 - \cos(2x)}{x^2}$$
---
2. $$\lim_{{x \to \infty}} \frac{3x^2 - 2x + 1}{2x^2 + x - 1}$$
---
3. $$\lim_{{x \to 0}} x \cdot \ln(1 + \sin(\frac{1}{x}))$$


Resolução algébrica:
--
1.
$
\lim_{{x \to 0}} \frac{{1 - \cos(2x)}}{{x^2}} 
$
2. Aplicação da identidade trigonométrica $1−cos⁡(2x)=2sin⁡²(x)$

  $
  \lim_{{x \to 0}} \frac{{2\sin^2(x)}}{{x^2}}
  $
---
3. Simplificação utilizando a propriedade $\lim_{{x \to 0}} \frac{{\sin(x)}}{{x}} = 1
$

  $
  2 \cdot \lim_{{x \to 0}} \frac{{\sin^2(x)}}{{x^2}}
  $
---
4. Agora, vamos aplicar a propriedade $\lim_{{x \to 0}} \frac{{f(x)}}{{g(x)}} = \frac{{\lim_{{x \to 0}} f(x)}}{{\lim_{{x \to 0}} g(x)}}$

  $2 \cdot \frac{{\lim_{{x \to 0}} \sin^2(x)}}{{\lim_{{x \to 0}} x^2}}$
---
5.O limite lim ⁡x→0 sin⁡²(x) é igual a 0, uma vez que sin⁡²(x) tende a 0 quando x tende a 0. O limite lim ⁡x→0 x² também é igual a 0. Portanto, temos:

  $2 \cdot \frac{{0}}{{0}}$
---
6. Regra de L'Hôpital

$2 \cdot \lim_{{x \to 0}} \frac{{\frac{{d}}{{dx}}(\sin^2(x))}}{{\frac{{d}}{{dx}}(x^2)}}
$
---
7. A derivada de sin⁡²(x) em relação a x é 2 sin⁡(x)cos⁡(x), e a derivada de x² em relação a x é 2x. Portanto, temos:

$2 \cdot \lim_{{x \to 0}} \frac{{2\sin(x)\cos(x)}}{{2x}}
$
---
8. Agora, podemos substituir x por 0 e obter o valor do limite:

$2 \cdot \frac{{\sin(0)\cos(0)}}{{0}}
$
---
9. Como sin⁡(0)=0 e cos⁡(0)=1, temos:

$2 \cdot \frac{{0 \cdot 1}}{{0}} = 2 \cdot \frac{{0}}{{0}}
$
---
10. L'Hôpital

$2 \cdot \lim_{{x \to 0}} \frac{{\frac{{d}}{{dx}}(\sin(x)\cos(x))}}{{\frac{{d}}{{dx}}(x)}}
$
---
11. A derivada de sin⁡(x)cos⁡(x) em relação a x é cos⁡²(x)−sin⁡²(x), e a derivada de x em relação a x é 1. Portanto, temos:
$2 \cdot \lim_{{x \to 0}} \frac{{\cos^2(x) - \sin^2(x)}}{{1}}
$
---
12. Podemos substituir x por 0 novamente:
$2 \cdot (\cos^2(0) - \sin^2(0)) = 2 \cdot (1 - 0) = 2
$
Portanto, o valor do limite $\lim_{{x \to 0}} \frac{{1 - \cos(2x)}}{{x^2}}$ é 2

```python
#Gráfico 1

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (1 - np.cos(2*x)) / (x**2)

intervalo = np.linspace(-0.2, 0.2, 100)

h = 0.01
x_esq = np.arange(-0.2, 0, h)
y_esq = f(x_esq)

x_dir = np.arange(0, 0.2, h)
y_dir = f(x_dir)

plt.plot(intervalo, f(intervalo), label='f(x)')
plt.plot(x_esq, y_esq, 'ro', label='Aproximação pela Esquerda')
plt.plot(x_dir, y_dir, 'bo', label='Aproximação pela Direita')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Gráfico do Limite')
plt.grid(True)
plt.show()

print("Tabela de Aproximação:")
print(" x \t\t f(x) (Esquerda) \t f(x) (Direita)")
for i in range(len(x_esq)):
    print(f"{x_esq[i]:.4f} \t {y_esq[i]:.4f} \t\t {y_dir[i]:.4f}")
```

Resolução algébrica 2:

1. L'Hôpital

$\lim_{{x \to \infty}} \frac{{\frac{{d}}{{dx}}(3x^2 - 2x + 1)}}{{\frac{{d}}{{dx}}(2x^2 + x - 1)}}
$

---

2. A derivada de 3x²−2x+1 em relação a xx é 6x−2, e a derivada de 2x²+x−1 em relação a x é 4x+1. Portanto, temos:

$
\lim_{{x \to \infty}} \frac{{6x - 2}}{{4x + 1}}
$

---

3. Agora, podemos avaliar o limite substituindo x por ∞:

$
\frac{{\lim_{{x \to \infty}} (6x - 2)}}{{\lim_{{x \to \infty}} (4x + 1)}}
$

---

4. À medida que x tende ao infinito, tanto 6x−2 quanto 4x+1 também tendem ao infinito. Portanto, temos:

$\frac{{\infty}}{{\infty}}
$

---

5. L'Hôpital

$
\lim_{{x \to \infty}} \frac{{\frac{{d}}{{dx}}(6x - 2)}}{{\frac{{d}}{{dx}}(4x + 1)}}
$

---

6. A derivada de 6x−2 em relação a x é 6, e a derivada de 4x+1 em relação a x é 4. Portanto, temos:

$
\lim_{{x \to \infty}} \frac{{6}}{{4}}
$

O valor do limite é 6/4 = 3/2

```python
#Gráfico 2

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (3*x**2 - 2*x + 1) / (2*x**2 + x - 1)

intervalo = np.linspace(-10, 10, 100)

h = 0.1
x_esq = np.arange(-10, 0, h)
y_esq = f(x_esq)

x_dir = np.arange(0, 10, h)
y_dir = f(x_dir)

plt.plot(intervalo, f(intervalo), label='f(x)')
plt.plot(x_esq, y_esq, 'ro', label='Aproximação pela Esquerda')
plt.plot(x_dir, y_dir, 'bo', label='Aproximação pela Direita')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Gráfico do Limite')
plt.grid(True)
plt.show()

print("Tabela de Aproximação:")
print(" x \t\t f(x) (Esquerda) \t f(x) (Direita)")
for i in range(len(x_esq)):
    print(f"{x_esq[i]:.1f} \t\t {y_esq[i]:.4f} \t\t {y_dir[i]:.4f}")
```

Resolução algébrica 3

1. A função ln⁡(1+sin⁡(1x)) é contínua para valores de 1/x​ no intervalo −∞ < 1/x < ∞, exceto quando 1+sin⁡(1/x)≤0. No entanto, como sin⁡(1/x) está limitado entre -1 e 1, 1+sin⁡(1/x) nunca será menor ou igual a 0 para valores suficientemente pequenos de x.

Portanto, para calcular o limite, podemos usar a propriedade lim⁡x→0ln⁡(1+x)=ln⁡(1+0)=ln⁡(1)=0. Substituindo essa propriedade na expressão do limite, temos:

$
\lim_{{x \to 0}} x \cdot \ln(1+\sin(\frac{1}{x})) = \lim_{{x \to 0}} x \cdot 0 = 0
$
Assim, o valor do limite é 0.

```python
#Gráfico 3
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x * np.log(1 + np.sin(1/x))

intervalo = np.linspace(-0.5, 0.5, 100)

h = 0.01
x_esq = np.arange(-0.5, 0, h)
y_esq = f(x_esq)

x_dir = np.arange(0, 0.5, h)
y_dir = f(x_dir)

plt.plot(intervalo, f(intervalo), label='f(x)')
plt.plot(x_esq, y_esq, 'ro', label='Aproximação pela Esquerda')
plt.plot(x_dir, y_dir, 'bo', label='Aproximação pela Direita')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Gráfico do Limite')
plt.grid(True)
plt.show()

print("Tabela de Aproximação:")
print(" x \t\t f(x) (Esquerda) \t f(x) (Direita)")
for i in range(len(x_esq)):
    print(f"{x_esq[i]:.2f} \t\t {y_esq[i]:.4f} \t\t {y_dir[i]:.4f}")
```

# Questão 2
Considerando a função definida por: 
A função descontínua $f(x)$ é definida como:

$$
f(x) = 
\begin{cases} 
\sin(x) & \text{se } -10 \leq x < -3 \\
\cos(x) & \text{se } -3 \leq x < 0 \\
x^2 & \text{se } 0 \leq x < 3 \\
e^{x-10} & \text{se } 3 \leq x \leq 10
\end{cases}
$$

Trace o gráfico desta função, marque os pontos de descontinuidade, caso existam estes pontos, com um círculo vermelho, anote estes pontos. Não se esqueça de colocar uma legenda que permita o entendimento do seu gráfico. 

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    if -10 <= x < -3:
        return np.sin(x)
    elif -3 <= x < 0:
        return np.cos(x)
    elif 0 <= x < 3:
        return x**2
    elif 3 <= x <= 10:
        return np.exp(x) - 10

x = np.linspace(-10, 10, 1000)
y = [f(xi) for xi in x]

descontinuidades = [-3, 0, 3]

plt.plot(x, y, label='f(x)')
plt.scatter(descontinuidades, [f(xi) for xi in descontinuidades], color='red', marker='o', label='Descontinuidades')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Gráfico da Função Descontínua')
plt.grid(True)
plt.show()
```
