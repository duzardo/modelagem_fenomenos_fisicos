# Questão 1: 
Um elemento radioativo tem uma meia-vida de 5 anos. Se inicialmente existem 500 gramas desse elemento, faça um gráfico que mostre como a quantidade desse elemento varia com o tempo e determine quantos gramas restarão após 15 anos. Utilize a equação $A = A0 * e^{-kt}$, onde $A0$ é a quantidade inicial do elemento, $k$ é a constante de decaimento (em 1/anos) e $t$ é o tempo decorrido em anos.



```python
import matplotlib.pyplot as plt
import numpy as np

A0 = 500
k = np.log(2)/5
t = np.arange(0, 15.1, 0.1)

A = A0 * np.exp(-k*t)

plt.plot(t, A)
plt.xlabel("Anos")
plt.ylabel("Quantidade")
plt.show()

A15 = A0 * np.exp(-k*15)
print(f"Restarão {A15:.2f} gramas após 15 anos.")
```

# Questão 2
Usando, a seu critério as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $f(x)=2^x + 3^x + 25$

2. $f(x)=\log_2(x) - \log_2(3) - 1$

```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: 2*x + 3*x + 25
x = np.linspace(-10, 10, 100)
y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('$2x + 3x + 25$')
plt.title('Função Crescente!')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: np.log2(x) - np.log2(3) - 1
x = np.linspace(0.01, 10, 100)
y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('$\log_2(x) - \log_2(3) - 1$')
plt.title('Função decrescente!')
plt.show()
```

# Questão 3
Usando o Sympy, ou qualquer outro recurso disponível na internet, escreva, com qualidade matemática o passo-a-passo algébrico que leva a determinação dos seguintes limites. Ao final, para cada uma das equações você deverá traçar um gráfico da função. 

1. $\lim_{x \to 2} \frac{x^3 - 8}{x - 2}$

2. $\lim_{x \to \infty} \frac{3x^2 + 2x - 1}{x^2 - 4x + 7}$



```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: (x**3 -8) / (x - 2)
g = lambda x:  x**2 + 2*x +4
x = np.linspace(1, 3.5, 100)

plt.plot(x, f(x), label='(x³ - 8) / (x - 2)')
plt.plot(x, g(x), label='x² + 2x + 4')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Limitando a função (x³ - 8) / (x - 2) em x = 2')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: (3*x**2 + 2*x - 1) / (x**2 - 4*x + 7)
g = lambda x: np.full_like(x, 3)
x = np.linspace(-10, 10, 100)

plt.plot(x, f(x), label='(3x² + 2x - 1) / (x² - 4x + 7)')
plt.plot(x, g(x), label='3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Limitando a função (3x² + 2x - 1) / (x² - 4x + 7) em x = ∞')
plt.legend()
plt.show()
```
