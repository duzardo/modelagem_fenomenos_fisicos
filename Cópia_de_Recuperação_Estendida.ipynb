# Questão 1 (vale 1 ponto)
Suponha que você esteja projetando uma ponte suspensa e precise determinar a deformação máxima que ocorrerá em um determinado ponto da estrutura. A partir de análises e medições preliminares, você sabe que a deformação em um ponto na posição x da ponte pode ser aproximada por uma função $f(x) = 3x^2 - 2x + 5$. Precisamos saber a deformação quando $x$ se aproxima do valor $2$.

Utilizando a Fórmula das diferenças finitas encontre os limites para cinco pontos à esquerda do ponto desejado e cinco pontos à direita do ponto desejado. Depois traçe um gráfico desta função anote o limite no ponto desejado e imprima os limites nos cinco pontos à esquerda e à direita.




## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 3*x**2 - 2*x + 5

def approx_derivative(f, x, h):
    return (f(x + h) - f(x)) / h

x = np.linspace(0, 4, 100)
y = f(x)

x_desejado = 2
y_desejado = f(x_desejado)

h = 0.1

x_esquerda = np.linspace(x_desejado - 0.5, x_desejado - 0.1, 5)
limits_esquerda = [approx_derivative(f, x, h) for x in x_esquerda]

x_direita = np.linspace(x_desejado + 0.1, x_desejado + 0.5, 5)
limits_direita = [approx_derivative(f, x, h) for x in x_direita]

plt.plot(x, y, label='f(x) = 3x² - 2x + 5')
plt.scatter(x_desejado, y_desejado, color='red', label='Ponto Desejado (x = 2)')
plt.scatter(x_esquerda, f(x_esquerda), color='green', label='Pontos à Esquerda')
plt.scatter(x_direita, f(x_direita), color='blue', label='Pontos à Direita')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função f(x) = 3x² - 2x + 5')
plt.grid(True)
plt.show()

print("Limites à esquerda:")
for i in range(len(x_esquerda)):
    print(f"No ponto x = {x_esquerda[i]:.1f}, o limite é {limits_esquerda[i]:.2f}")

print("\nLimites à direita:")
for i in range(len(x_direita)):
    print(f"No ponto x = {x_direita[i]:.1f}, o limite é {limits_direita[i]:.2f}")
```

# Exercício de Limites e Descontinuidades

Considere as seguintes funções:

1. $$ f(x) = \tan(x) $$
2. $$ g(x) = \frac{1}{x-5} $$

Para cada função, trace o gráfico no intervalo de 0 a 10 usando Python, Matplotlib e NumPy. a)Determine, se existem, ou não descontinuidade para estas funções no intervalo solicitado. b)Caso exista alguma descontinuidade encontre o limite em um ponto qualquer à direita e outro a esquerda do ponto onde existe a descontinuidade. c)Você precisa plotar os gráficos destas funções usando python, e apenas as bibliotecas matplotlib, numpy e sympy.


## Resposta

```python
#1
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.tan(x)

x = np.linspace(0, 10, 1000)

y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de f(x) = tan(x)')
plt.grid(True)
plt.show()

discontinuities = np.where(np.abs(np.diff(y)) > 10)[0] + 1

if len(discontinuities) == 0:
    print("Não há descontinuidades no intervalo de 0 a 10.")
else:
    print("Descontinuidade(s) encontrada(s) no intervalo de 0 a 10.")
    for i in discontinuities:
        limit_left = np.tan(x[i - 1])
        limit_right = np.tan(x[i])
        print(f"Descontinuidade em x = {x[i]}")
        print(f"Límite à esquerda: {limit_left}")
        print(f"Límite à direita: {limit_right}")
```

```python
#2
import numpy as np
import matplotlib.pyplot as plt

def g(x):
    return 1 / x - 5

x = np.linspace(0.01, 10, 1000)

y = g(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Gráfico de g(x) = 1/x - 5')
plt.grid(True)
plt.show()

discontinuities = np.where(np.abs(np.diff(y)) > 10)[0] + 1

if len(discontinuities) == 0:
    print("Não há descontinuidades no intervalo de 0 a 10.")
else:
    print("Descontinuidade(s) encontrada(s) no intervalo de 0 a 10.")
    for i in discontinuities:
        limit_left = 1 / x[i - 1] - 5
        limit_right = 1 / x[i] - 5
        print(f"Descontinuidade em x = {x[i]}")
        print(f"Límite à esquerda: {limit_left}")
        print(f"Límite à direita: {limit_right}")
```

#Questão 4 (vale 2 pontos)
Considere a função $f(x)= x^2 -1$. (a) Trace o gráfico desta função e, (b) usando a definição de derivada, encontre a inclinação da reta tangente nos pontos $(-1,0)$, $(0,-1)$ e $(2,3)$.

## Resposta Algébrica


1. Para o ponto (-1, 0):

\
\begin{align*}
f'(-1) &= \lim_{{h \to 0}} \frac{{f(-1 + h) - f(-1)}}{h} \\
&= \lim_{{h \to 0}} \frac{{(-1 + h)^2 - 1}}{h} \\
&= \lim_{{h \to 0}} \frac{{h^2 - 2h}}{h} \\
&= \lim_{{h \to 0}} (h - 2) \\
&= -2
\end{align*}


Portanto, a inclinação da reta tangente no ponto \((-1, 0)\) é -2.

2. Para o ponto (0, -1):


\begin{align*}
f'(0) &= \lim_{{h \to 0}} \frac{{f(0 + h) - f(0)}}{h} \\
&= \lim_{{h \to 0}} \frac{{(0 + h)^2 - 1}}{h} \\
&= \lim_{{h \to 0}} \frac{{h^2 - 1}}{h} \\
&= \lim_{{h \to 0}} (h - \frac{1}{h})
\end{align*}


A derivada em (0, -1) é indeterminada. No entanto, sabemos que a derivada da função $(f(x) = x^2 - 1)$ é $(f'(x) = 2x)$. Portanto, substituindo (x = 0) na derivada, obtemos:


$f'(0) = 2 \cdot 0 = 0$


Assim, a inclinação da reta tangente no ponto (0, -1) é 0.

3. Para o ponto (2, 3):


\begin{align*}
f'(2) &= \lim_{{h \to 0}} \frac{{f(2 + h) - f(2)}}{h} \\
&= \lim_{{h \to 0}} \frac{{(2 + h)^2 - 1}}{h} \\
&= \lim_{{h \to 0}} \frac{{h^2 + 4h + 3}}{h} \\
&= \lim_{{h \to 0}} (h + 4 + \frac{3}{h})
\end{align*}

A derivada em (2, 3) também é indeterminada. Usando a derivada conhecida $(f'(x) = 2x)$, substituindo (x = 2), obtemos:

$
f'(2) = 2 \cdot 2 = 4
$

Portanto, a inclinação da reta tangente no ponto (2, 3) é 4.

Resumindo as inclinações das retas tangentes nos pontos fornecidos:

(-1, 0): A inclinação da reta tangente é -2.
(0, -1): A inclinação da reta tangente é 0.
(2, 3): A inclinação da reta tangente é 4


## Resposta com código

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 1

x = np.linspace(-5, 5, 100)

y = f(x)

x_points = [-1, 0, 2]
y_points = [f(x) for x in x_points]
slopes = [-2, 0, 4]

plt.plot(x, y, label='f(x) = x^2 - 1')
plt.scatter(x_points, y_points, color='red', label='Pontos de inclinação')
for i, txt in enumerate(slopes):
    plt.annotate(txt, (x_points[i], y_points[i]), textcoords="offset points", xytext=(0,10), ha='center')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função f(x) = x^2 - 1')
plt.grid(True)
plt.legend()
plt.show()

```

# Questão 5 (vale 1 ponto)
Você é um engenheiro de produção em uma fábrica de vidro que produz garrafas. A velocidade de produção das garrafas $V$ varia ao longo do tempo devido ao desgaste das máquinas e a fadiga dos trabalhadores. Depois de um mês de análise, você descobriu que a velocidade de produção é dada pela função $V(t) = 50 - 4t^2$ garrafas por hora, onde $t$ é o tempo em horas desde o início do turno de trabalho.
Seu gerente quer saber se há um momento específico durante o turno de trabalho em que a produção de garrafas é mais eficiente.
a) Como você poderia ajudar seu gerente a encontrar essa informação?
b) Se este momento existe, que momento é este?
c) Trace o gráfico da sua função usando apenas python, matplotlib, numpy e sympy e, caso este momento exista, marque este ponto no seu gráfico.

## Resposta  
a)
\begin{align*}
V(t) &= 50 - 4t^2
\end{align*}
Calculamos a derivada aplicando a regra da potência e a regra da constante
\begin{align*}
\frac{dV(t)}{dt} &= \frac{d}{dt}(50) - \frac{d}{dt}(4t^2)
\end{align*}
A derivada constante é 0, então ficamos com:
\begin{align*}
\frac{dV(t)}{dt} &= 0 - 4\frac{d}{dt}(t^2) \\
\\
\end{align*}
Aplicando a regra da potência, temos:
\begin{align*}
\frac{dV(t)}{dt} &= -4(2t) \\
\end{align*}
Simplificando a expressão:
\begin{align*}
\frac{dV(t)}{dt} &= -8t \\
\end{align*}

b)
Para encontrar o momento específico durante o turno de trabalho em que a produção de garrafas é mais eficiente, precisamos determinar o valor máximo da função de velocidade de produção (V(t)).

A função de velocidade de produção é dada por:

$V(t) = 50 - 4t^2$

Para encontrar o valor máximo de (V(t)), precisamos derivar a função em relação a (t) e igualar a derivada a zero, pois o ponto de máximo ocorre quando a taxa de variação é igual a zero.

Vamos calcular a derivada de (V(t)):

$V(t) = \frac{dV}{dt} = -8t$

Agora, igualamos a derivada a zero para encontrar o valor de (t) onde a produção é mais eficiente:

$-8t = 0$

$t = 0$

Portanto, o momento mais eficiente durante o turno de trabalho é no início do turno, quando t = 0 horas.

Dessa forma, o valor máximo da função de velocidade de produção ocorre no início do turno, ou seja, a produção é mais eficiente no início do turno de trabalho.





```python
#Gráfico
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def V(t):
    return 50 - 4 * t**2

t = np.linspace(0, 6, 100)
v = V(t)


t_max = sp.Symbol('t')
v_max = V(t_max)
t_max_value = sp.solve(sp.diff(v_max, t_max), t_max)[0]

plt.plot(t, v, label='V(t) = 50 - 4t^2')

plt.plot(t_max_value, V(t_max_value), 'ro', label='Momento mais eficiente')

plt.xlabel('Tempo (horas)')
plt.ylabel('Produção de garrafas por hora')

plt.legend()

plt.show()
```

# Questão 6 (vale 1 ponto)
Considere a função $f(x) = x^3 - 10x^2 + sin(x)$. Usando apenas o Python, numpy, sympy e matplotlib, trace o gráfico desta função e a reta tangente ao ponto $x=4$.

## Resposta

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = x*3 - 10*x*2 + sp.sin(x)

f_prime = f.diff(x)

f_func = sp.lambdify(x, f)
f_prime_func = sp.lambdify(x, f_prime)

x_vals = np.linspace(0, 6, 100)
y_vals = f_func(x_vals)
tangent_vals = f_func(4) + f_prime_func(4) * (x_vals - 4)

plt.plot(x_vals, y_vals, label='f(x)')
plt.plot(x_vals, tangent_vals, label='Tangente em x=4')
plt.scatter([4], [f_func(4)], color='red', label='Ponto (4, f(4))')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

#Questão 7 (vale 1 ponto)
Considere um fenômeno físico em que a posição de um objeto em movimento varia com o tempo. A taxa de variação da posição em relação ao tempo possui um componente trigonométrico e é dada pela função $f(t) = A * cos(t)$, onde $A$ é uma constante que determina a amplitude da variação. Sua tarefa é plotar o gráfico da função da taxa de variação e da função posição deste objeto considerando que $A$ é um número inteiro que varia entre 2 e 5.

## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2*np.pi, 100)

A_values = range(2, 6)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Função da taxa de variação')
plt.xlabel('t')
plt.ylabel('f(t)')

for A in A_values:
    f = A * np.cos(t)
    plt.plot(t, f, label='A = {}'.format(A))

plt.legend()

plt.subplot(1, 2, 2)
plt.title('Função posição')
plt.xlabel('t')
plt.ylabel('x(t)')

for A in A_values:
    x = A * np.sin(t)
    plt.plot(t, x, label='A = {}'.format(A))

plt.legend()
plt.tight_layout()
plt.show()
```

#Questão 8 (vale 2 pontos)
Calcule a área entre as curvas $f(x) = -x^2$ e $g(x) = -x$, apenas no quarto quadrante do plano cartesiano, usando apenas álgebra e garantindo que todos os passos algébricos necessários estejam explicitados em notação matemática em uma célula de textos deste carderno de tarefas usando latex. Usando o python, matplotlib, numpy e sympy crie um gráfico destas funções sombreando a área entre elas em azul claro.

## Resposta



\\begin{align*}f(x) &= -x^2 \\g(x) &= -x\end{align*}

Passo 1: Encontrar os pontos de interseção
$
-x^2 = -x
$

Passo 2: Resolver a equação
$
-x^2 + x = 0
$

Passo 3: Fatorar a equação
$
x(-x + 1) = 0
$

Passo 4: Encontrar os valores de x
Os pontos de intercessão $x = 0$ e $x = 1.$

Passo 5: Calcular a área
Área = $\int_{0}^{1} (g(x) - f(x)) \, dx$

Passo 6: Integração
Área = $\int_{0}^{1} (-x + x^2) \, dx$

Passo 7: Avaliar a integral nos limites
Área = $\left[ -\frac{{x^2}}{2} + \frac{{x^3}}{3} \right]_{0}^{1}$

Área = $\left( -\frac{1}{2} + \frac{1}{3} \right) - \left( -\frac{0^2}{2} + \frac{0^3}{3} \right)$

Área = $-\frac{1}{6}$

Portanto, a área entre as curvas $(f(x) = -x^2)$ e $(g(x) = -x)$ no quarto quadrante do plano cartesiano é $(-\frac{1}{6})$.


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x = sp.symbols('x')

f = -x**2
g = -x

f_np = sp.lambdify(x, f, 'numpy')
g_np = sp.lambdify(x, g, 'numpy')

x_vals = np.linspace(0, 1, 100)

f_vals = f_np(x_vals)
g_vals = g_np(x_vals)

plt.plot(x_vals, f_vals, label='f(x)=-x^2')
plt.plot(x_vals, g_vals, label='g(x)=-x')
plt.fill_between(x_vals, f_vals, g_vals, where=(f_vals > g_vals), color='lightblue')


plt.title('Área entre as Curvas f(x)=-x^2 e g(x)=-x')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.show()
```
